# Example 3. Figure 5 "Hubble's law"

## 1. Make data list for plots
- For making figure 5, I will not import data files. Instead, I will make new lists in here.
- Data of velocity-distance relation among extra-galactic nebulae is given (by professor) as v, d list respectively.
-  Hubble constant is assumed as 500km/s/Mpc
- To express the equation v = H0 * d, set another list of x(distance), y(recession velocity)  which will be the straight line.
  - Because the x, which is distance is too large, 
  
   use unit Mpc and set new variable xMpc for x coordinate.

```python
#Example 3. (Figure 5) Hubble's law 

# use two lists for d and v (given)
# use green filled triangles for a scatter plot
# draw a black solid line for H0= 500km/s/Mpc
# add axis labels 
# add a figure title
# add a text "H0=500 km/s/Mpc"
################################################

# Hubble's original data (15 data points are read from Figure 1, 1929 paper by Hubble)
#read a data file  --> time versus simulated GW signal for GW150914

d=[ 27888.44621513982, 35856.57370517962, 215139.44223107595, 266932.27091633494, 501992.0318725102, 637450.1992031876,900398.406374502, 804780.8764940242, 896414.3426294823,908366.5338645419,1103585.657370518, 
   1107569.7211155382, 1394422.310756972, 1713147.4103585659, 2007968.12749004] # distance to a galaxy in Mpc
   
v=[ -22.5816733067727, 37.45019920318714, 38.16733067729069, 6.37450199203181, 423.31474103585656, 327.85657370517936, 216.9083665338644, 380.52589641434247, 436.89243027888426, 616.9402390438247, 577.7211155378485
   ,737.7370517928287, 626.8844621513944, 1060.1593625498008, 541.3386454183266] # recession velocity of a galaxy in km/s

# Assume H0=500 km/s/Mpc and draw a solid line overlaid on the triangle points
H0=500 # in km/s/Mpc Hubble's original value of H0


# set an arbitrary list [x] to draw a line

# line plotting
x= [0, 2]  # x in Mpc  (단위 제외함)
y= [i * H0 for i in x] # y in km/s <--- y = H0 * x ==> for loop 돌아간 후 y=[0, 1000]
xMpc=[i*1e6 for i in x] # multiply 1e6 for unit consistency with triangle points, use this for plotting a line

```

## 2. Drawing plot

####1st plot. scatter plot with d against v
  - 'g^' means 'use green triangle marker'
  - set markersize as 16

####2nd plot. solid line plot with xMpc against y with the given H0.
  - using command color = 'black', linestyle = '-', linewidth = '4'

  to use solid black line with width 4 as line plot.


###Adding label, text, title
  - plt.xlabel, plt.text, plt.title

```python

### make a figure ###


# set a font size to be 22
plt.rcParams.update({'font.size': 22})

# Set picture size 
fig = plt.figure(figsize=(8,6))

# set styles for points for a scatter plot with d versus v. use green triangles and markersize="16"
plt.plot(d, v, 'g^', markersize=16)

# set line style, a line plot with xMpc versus y with the given H0, use a black solid line and linewidth='4'
plt.plot(xMpc, y, color ="black", linestyle = '-', linewidth = '4')

# Set labels of x-axis (distance (Mpc)), y-axis (recession velocity (km/s))  
plt.xlabel("distance (Mpc)")
plt.ylabel("recession velocity (km/s)")


# add a text in the figure
plt.text(150, 950, '$H_0=500$ km/s/Mpc', {'color': 'black', 'fontsize': 20})
# plt.text(x,y) <-- x,y are in data coordinates


# Set figure title  (Hubble's law)
plt.title("Hubble's law")

#show the figure on the screen
plt.show()
```
